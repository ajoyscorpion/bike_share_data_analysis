### Data Cleaning

```python
trip_data['date']=pd.to_datetime(trip_data['started_at'])
trip_data['month'] = trip_data['date'].dt.strftime('%b')
trip_data['day'] = trip_data['date'].dt.strftime('%d')
trip_data['year'] = trip_data['date'].dt.strftime('%Y')
trip_data['day_of_week'] = trip_data['date'].dt.strftime('%A')
```

```python
trip_data.head()
```

#### Removing NA's

```python
trip_data.isna().sum()
```

```python
trip_data = trip_data.dropna()
```

```python
trip_data.isna().sum()
```

```python
trip_data_no_duplicates = trip_data.drop_duplicates(subset=['ride_id'])
```

```python
print("Removed", len(trip_data) - len(trip_data_no_duplicates), "duplicate rows")
```

```python
trip_data.shape[0]
```

```python
trip_data_no_duplicates.shape[0]
```

### Data Manipulation

```python
trip_data_no_duplicates['started_at']=pd.to_datetime(trip_data_no_duplicates['started_at'])
trip_data_no_duplicates['ended_at']=pd.to_datetime(trip_data_no_duplicates['ended_at'])
```

```python
trip_data_no_duplicates
```

```python
trip_data_no_duplicates['ride_length'] = (trip_data_no_duplicates['ended_at'] - trip_data_no_duplicates['started_at']).dt.total_seconds() / 60
```

```python
trip_data_no_duplicates.info()
```

```python
trip_data_no_duplicates
```

#### filtering out trips with a ride length less than 0.

```python
trip_data_v2 = trip_data_no_duplicates[trip_data_no_duplicates['ride_length']>=0]
```

```python
trip_data_v2
```

#### Number of members vs casual riders

```python
rider_type = trip_data_v2['member_casual'].value_counts()
```

```python
rider_type
```

#### Trip Data Statistics

```python
import numpy as np
```

```python
trip_data_stat = trip_data_v2.groupby('member_casual')['ride_length'].agg(
    average_ride_length = np.mean,
    standard_deviation = np.std,
    median_ride_length = np.median,
    min_ride_length = np.min,
    max_ride_length = np.max
).reset_index()
```

```python
trip_data_stat
```

#### Determining the most popular day by rider type

```python
weekday_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
trip_data_v2['day_of_week'] = pd.Categorical(trip_data_v2['day_of_week'], categories=weekday_order, ordered=True)
```

```python
result = trip_data_v2.groupby(['member_casual','day_of_week']).agg(
    rider_type_total=('ride_id','count'),
    average_ride_length=('ride_length','mean')
)
```

```python
result = result.sort_values(by=['member_casual','day_of_week'])
```

```python
result
```

#### Popular month

```python
popular_month=trip_data_v2.groupby('month').agg(
    number_of_rides=('ride_id','count'),
    average_duration=('ride_length','mean')
).reset_index()
```

```python
popular_month = popular_month.sort_values(by='number_of_rides',ascending=False)
```

```python
popular_month
```

#### Determine the most popular start station

```python
station_popular = trip_data_v2['start_station_name'].mode().iloc[0]
```

```python
station_popular
```

#### Popular Start Station for Member

```python
popular_start_station_member = trip_data_v2[trip_data_v2["member_casual"]=="member"]\
    .groupby('start_station_name')\
    .size()\
    .reset_index(name='number_of_starts')\
    .sort_values(by='number_of_starts', ascending=False) \
    .head()
```

```python
popular_start_station_member
```

#### Popular Start Station for Casual

```python
popular_start_station_casual = trip_data_v2[trip_data_v2["member_casual"]=="casual"]\
    .groupby('start_station_name')\
    .size()\
    .reset_index(name='number_of_starts')\
    .sort_values(by='number_of_starts', ascending=False) \
    .head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Rider types

```python
plt.figure(figsize=(6,6))
sns.countplot(x="member_casual",data=trip_data_v2,palette='viridis')
plt.title('Distribution of Rider Types')
plt.xlabel('Rider Type')
plt.ylabel('Total Count')
plt.show()
```

#### Rider types ride duration

```python
rider_type_average_duration = trip_data_v2.groupby('member_casual')['ride_length'].mean().reset_index()
rider_type_average_duration.columns = ['member_casual', 'average_ride_length']
rider_type_average_duration
```

```python
plt.figure(figsize=(6, 6))
sns.barplot(x='member_casual', y='average_ride_length', data=rider_type_average_duration, palette='viridis')
plt.title('Rider Types Ride Duration')
plt.xlabel('Member Type')
plt.ylabel('Average Ride Duration')
plt.show()
```

#### The usage by members and casual riders by the weekday

```python
average_duration = trip_data_v2.groupby(['member_casual', 'day_of_week'])['ride_length'].mean().reset_index()
average_duration.columns = ['member_casual', 'day_of_week' ,'ride_length']
plt.figure(figsize=(7, 6))
sns.barplot(x='day_of_week', y='ride_length', hue='member_casual', data=average_duration, palette='viridis')
plt.xlabel('Day of Week')
plt.ylabel('Average Duration')
plt.title('Average Duration by Member Type and Day of Week')
plt.show()
average_duration
```

#### The number of trips by members and casual riders by the weekday

```python
number_of_rides = trip_data_v2.groupby(['member_casual', 'day_of_week'])['ride_length'].count().reset_index()
number_of_rides.columns = ['member_casual', 'day_of_week' ,'ride_length']
plt.figure(figsize=(7, 6))
sns.barplot(x='day_of_week', y='ride_length', hue='member_casual', data=number_of_rides, palette='viridis')
plt.xlabel('Day of Week')
plt.ylabel('Number of Rides')
plt.title('Number of Rides by Member Type and Day of Week')
plt.show()
number_of_rides
```

#### The usage by members and casual riders by the month

```python
months_order = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
trip_data_v2['month'] = pd.Categorical(trip_data_v2['month'], categories=months_order, ordered=True)

rides_duration_by_month = trip_data_v2.groupby(['member_casual', 'month'])[['ride_length']].agg(['count', 'mean']).reset_index()

plt.figure(figsize=(6, 6))
sns.barplot(x='month', y=('ride_length', 'mean'), hue='member_casual', data=rides_duration_by_month, palette='viridis')

plt.xlabel('Month')
plt.ylabel('Average Duration')
plt.title('Average Duration of Rides by Member Type and Month')

plt.show()
rides_duration_by_month
```

#### The number of trips by members and casual riders by the month

```python
rides_by_month = trip_data_v2.groupby(['member_casual', 'month'])[['ride_id']].count().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='month', y='ride_id', hue='member_casual', data=rides_by_month, palette='viridis')

plt.xlabel('Month')
plt.ylabel('Number of Rides')
plt.title('Number of Rides by Member Type and Month')
plt.show()

rides_by_month
```
